```python
import pandas as pd
import numpy as np

import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Q1 : Version Pandas

pd.__version__
print("la version pandas est : ", pd.__version__)
```

```python
# Importation des données

df = pd.read_csv('https://raw.githubusercontent.com/alexeygrigorev/datasets/master/car_fuel_efficiency.csv')
df.head(10)
```

```python
# Q2 : Nombre d’enregistrements

print("le nombre d'enregistrements est : ", len(df))
```

```python
# Q3 : Types de combustibles

df['fuel_type'].unique()
print("nous avons ",len(df['fuel_type'].unique()), "types de combustibles différents")
print("les types de combustibles sont : ", df['fuel_type'].unique())
```

```python
# Q4 : Valeurs manquantes

nb = 0
list(df.isna().sum())
for i in range(len(list(df.isna().sum()))):
    if list(df.isna().sum())[i] != 0:
      nb = nb + 1
print("il y a ", nb, "variables avec des manquantes dans la colonne")
```

```python
#Q5 : Efficacité énergétique maximale

asia_cars = df[df['origin'] == 'Asia']
max_fuel_efficiency_asia = asia_cars['fuel_efficiency_mpg'].max()
print("L'efficacité énergétique maximale pour les voitures d'Asie est : ",round(max_fuel_efficiency_asia,2))
```

```python
# Q6 : Valeur médiane de la puissance

mediane_depart = df['horsepower'].median()
print("La valeur médiane de la puissance est : ",round(mediane_depart,2))

valeur_frequente = df['horsepower'].mode()[0]
print("La valeur la plus fréquente de la puissance est : ",round(valeur_frequente,2))

df['horsepower'].fillna(valeur_frequente, inplace=True)
mediane_remplacee = df['horsepower'].median()
print("La valeur médiane de la puissance après remplacement est : ",round(mediane_remplacee,2))
```

```python
# Q7 : Somme des poids

# Selection des voitures Asiatiques
asia_cars = df[df['origin'] == 'Asia']

# Sélectionnez uniquement les colonnes et vehicle_weightmodel_year
asia_cars_filtered = asia_cars[['vehicle_weight', 'model_year']]

# Selection des 7 premières voitures
asia_cars_7 = asia_cars_filtered.head(7)

# Récupérez le tableau NumPy sous-jacent. Appelons-le .X
X = asia_cars_7.values

# Calculez la multiplication matrice-matrice entre la transposée de et . Pour obtenir la transposition, utilisez . Appelons le résultat .XXX.TXTX
XTX = X.T.dot(X)

# Inverser XTX
XTX_inv = np.linalg.inv(XTX)

# Créez un tableau avec des valeurs .y[1100, 1300, 800, 900, 1000, 1100, 1200]
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])

# Multipliez l’inverse de par la transposée de , puis multipliez le résultat par . Appelez le résultat .XTXXyw
w = XTX_inv.dot(X.T).dot(y)

# Affichez les valeurs de w
print(w)

# Quelle est la somme de tous les éléments du résultat ?
print("La somme de tous les éléments du résultat est : ",round(w.sum(),4))
```
